```python
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {}
header = ['V', 'J', 'Pe', 'Ct', 'Cp', 'PWR', 'Torque', 'Thrust']        
with open('PER3_6x3.dat.txt', 'r') as f:
    for line in f.readlines():
        
        match_rpm = re.match('.*PROP.*RPM =\s*([0-9]+)', line)
        if match_rpm is not None:
            rpm = int(match_rpm.group(1))
            data[rpm] = []
            continue

        match_data = re.match('(\s*[-]?\d+\.\d*\s*)+', line)
        if match_data is not None:
            line_data = [float(v) for v in match_data.group(0).split()]
            data[rpm].append(line_data)
            if len(line_data) != 8:
                raise IOError(line)

    for k in data.keys():
        data[k] = pd.DataFrame(data[k], columns=header)
```

```python
for k in [1000, 6000, 10000, 20000]:
    plt.plot(data[k].J, data[k].Ct, label=k)
    plt.legend()
```

```python
ct_coef = np.polyfit(data[6000].J, data[6000].Ct, 4)
J = np.linspace(0,1)
plt.plot(J, np.polyval(ct_coef, J))
plt.plot(data[6000].J, data[6000].Ct, '.')
plt.xlabel('J')
plt.ylabel('Ct')
print(np.flip(ct_coef))
```

```python
cp_coef = np.polyfit(data[6000].J, data[6000].Cp, 4)
J = np.linspace(0,1)
plt.plot(J, np.polyval(cp_coef, J))
plt.plot(data[6000].J, data[6000].Cp, '.')
plt.xlabel('J')
plt.ylabel('Cp')
print(np.flip(cp_coef))
```
